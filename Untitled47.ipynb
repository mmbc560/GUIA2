```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

# -----------------------------
# Ejercicio: visualizar la metodología completa de la Guía 2
# -----------------------------

# Pasos de la metodología
pasos = [
    "1. Identificación\nde riesgos",
    "2. Identificación de\nmacroprocesos",
    "3. Evaluación de\nmadurez",
    "4. Priorización\nde procesos",
    "5. Caracterización\nde procesos críticos",
    "6. Análisis y\nconclusiones"
]

# Colores por etapa
colores = [
    "#ff6b6b",  # rojo suave
    "#4dabf7",  # azul
    "#ffd43b",  # amarillo
    "#69db7c",  # verde
    "#b197fc",  # morado
    "#ffa94d"   # naranja
]

# Posiciones horizontales
x_positions = [0.5, 2.4, 4.3, 6.2, 8.5, 10.8]
y = 0.5

# Crear figura
fig, ax = plt.subplots(figsize=(16, 4))
ax.set_xlim(0, 12)
ax.set_ylim(0, 1.2)
ax.axis("off")

# Título
ax.text(
    6, 1.05,
    "Metodología completa de la Guía 2",
    ha="center", va="center",
    fontsize=20, fontweight="bold"
)

# Dibujar cajas y flechas
box_width = 1.5
box_height = 0.35

for i, (x, paso, color) in enumerate(zip(x_positions, pasos, colores)):
    # Caja redondeada
    caja = FancyBboxPatch(
        (x - box_width / 2, y - box_height / 2),
        box_width,
        box_height,
        boxstyle="round,pad=0.03,rounding_size=0.08",
        linewidth=2,
        edgecolor="black",
        facecolor=color,
        alpha=0.95
    )
    ax.add_patch(caja)

    # Texto dentro de la caja
    ax.text(
        x, y,
        paso,
        ha="center", va="center",
        fontsize=11, fontweight="bold", color="black"
    )

    # Flecha hacia el siguiente paso
    if i < len(x_positions) - 1:
        flecha = FancyArrowPatch(
            (x + box_width / 2, y),
            (x_positions[i + 1] - box_width / 2, y),
            arrowstyle="->",
            mutation_scale=18,
            linewidth=2.5,
            color="gray"
        )
        ax.add_patch(flecha)

# Texto inferior explicativo
ax.text(
    6, 0.1,
    "Secuencia lógica del trabajo: primero se identifican riesgos, luego se analizan los procesos,\n"
    "se evalúa su madurez, se priorizan, se caracterizan y finalmente se plantean conclusiones.",
    ha="center", va="center",
    fontsize=11
)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# -----------------------------
# Ejercicio: identificación de riesgos de la empresa
# -----------------------------

# Riesgos del proceso
riesgos = [
    "Retrasos en entregas",
    "Caída del sistema",
    "Errores en pedidos",
    "Falta de inventario",
    "Rotación de personal"
]

# Probabilidad (escala 1 a 5)
probabilidad = [5, 4, 3, 4, 2]

# Impacto (escala 1 a 5)
impacto = [4, 5, 3, 5, 3]

# Cálculo de criticidad
criticidad = [p * i for p, i in zip(probabilidad, impacto)]

# Función para asignar color según criticidad
def asignar_color(valor):
    if valor <= 5:
        return "green"
    elif valor <= 10:
        return "yellow"
    elif valor <= 15:
        return "orange"
    else:
        return "red"

colores = [asignar_color(valor) for valor in criticidad]

# Crear gráfica
plt.figure(figsize=(12, 6))
bars = plt.bar(riesgos, criticidad, color=colores, edgecolor="black")

# Etiquetas encima de cada barra
for bar, valor in zip(bars, criticidad):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.3,
        str(valor),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold"
    )

# Títulos y ejes
plt.title("Identificación de riesgos de la empresa", fontsize=16, fontweight="bold")
plt.xlabel("Riesgos", fontsize=12)
plt.ylabel("Nivel de criticidad (Probabilidad x Impacto)", fontsize=12)
plt.ylim(0, max(criticidad) + 3)
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Mostrar
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ======================================
# LOS ESTUDIANTES SOLO MODIFICAN ESTA PARTE
# ======================================

macroproceso_critico = "Logística"

datos_procesos = {
    "Proceso": [
        "Recepción de pedidos",
        "Alistamiento de pedidos",
        "Despacho",
        "Entrega final",
        "Gestión de devoluciones"
    ],
    "Impacto en el cliente": [4, 4, 5, 5, 3],
    "Nivel de riesgo": [3, 4, 5, 5, 3],
    "Desempeño actual": [3, 3, 2, 2, 3],   # 1 = muy bajo, 5 = muy alto
    "Frecuencia del problema": [3, 4, 5, 4, 2]
}

# ======================================
# NO MODIFICAR DESDE AQUÍ
# ======================================

df = pd.DataFrame(datos_procesos)

# Convertir desempeño en urgencia:
# desempeño bajo = mayor urgencia de intervención
df["Bajo desempeño (urgencia)"] = 6 - df["Desempeño actual"]

# Calcular puntaje total
df["Puntaje total"] = (
    df["Impacto en el cliente"] +
    df["Nivel de riesgo"] +
    df["Bajo desempeño (urgencia)"] +
    df["Frecuencia del problema"]
)

# Ordenar por prioridad
df = df.sort_values(by="Puntaje total", ascending=False).reset_index(drop=True)

# Identificar los dos procesos más críticos
df["Prioridad"] = ["Crítico 1", "Crítico 2"] + ["Seguimiento"] * (len(df) - 2)

# --------------------------------------
# FIGURA CON GRÁFICA Y TABLA
# --------------------------------------
fig = plt.figure(figsize=(15, 9))
gs = fig.add_gridspec(2, 1, height_ratios=[3, 1.8])

ax = fig.add_subplot(gs[0])
ax_tabla = fig.add_subplot(gs[1])

# -----------------------------
# GRÁFICA
# -----------------------------
x = np.arange(len(df))
y = df["Puntaje total"]

barras = ax.bar(x, y, edgecolor="black")

for i, barra in enumerate(barras):
    if i == 0:
        barra.set_color("#e74c3c")   # rojo
    elif i == 1:
        barra.set_color("#f39c12")   # naranja
    else:
        barra.set_color("#9ecae1")   # azul claro

ax.set_xticks(x)
ax.set_xticklabels(df["Proceso"], rotation=15, ha="right", fontsize=11)
ax.set_ylabel("Puntaje total de priorización", fontsize=12)
ax.set_xlabel(f"Procesos del macroproceso: {macroproceso_critico}", fontsize=12)
ax.set_title(f"Priorización de procesos dentro de {macroproceso_critico}", fontsize=16, fontweight="bold")
ax.grid(axis="y", linestyle="--", alpha=0.4)

# valores sobre barras
for i, valor in enumerate(y):
    ax.text(i, valor + 0.2, int(valor), ha="center", fontsize=11, fontweight="bold")

# resaltar los dos más críticos
for i in range(2):
    ax.text(i, y.iloc[i] + 1.0, df["Prioridad"].iloc[i],
            ha="center", fontsize=11, fontweight="bold")

# -----------------------------
# TABLA
# -----------------------------
ax_tabla.axis("off")

columnas_mostrar = [
    "Proceso",
    "Impacto en el cliente",
    "Nivel de riesgo",
    "Desempeño actual",
    "Bajo desempeño (urgencia)",
    "Frecuencia del problema",
    "Puntaje total",
    "Prioridad"
]

tabla = ax_tabla.table(
    cellText=df[columnas_mostrar].values,
    colLabels=columnas_mostrar,
    cellLoc="center",
    loc="center"
)

tabla.auto_set_font_size(False)
tabla.set_fontsize(9)
tabla.scale(1, 1.6)

for (fila, col), cell in tabla.get_celld().items():
    if fila == 0:
        cell.set_text_props(weight="bold")
        cell.set_facecolor("#d9eaf7")
    else:
        # colorear filas críticas
        if fila == 1:
            cell.set_facecolor("#f8d7da")  # rojo suave
        elif fila == 2:
            cell.set_facecolor("#fdebd0")  # naranja suave
        else:
            cell.set_facecolor("#f8f8f8")

plt.tight_layout()
plt.show()
```
