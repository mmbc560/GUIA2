```python
import matplotlib.pyplot as plt

# Datos del proceso
etapas = ['Recepción', 'Revisión', 'Aprobación', 'Empaque', 'Despacho']
tiempos = [5, 8, 12, 6, 7]

# Identificar cuello de botella
cuello = max(tiempos)
indice_cuello = tiempos.index(cuello)

# Gráfica
colores = ['skyblue'] * len(etapas)
colores[indice_cuello] = 'red'

plt.figure(figsize=(10,6))
barras = plt.bar(etapas, tiempos, color=colores, edgecolor='black')
plt.title('Identificación del Cuello de Botella', fontsize=16)
plt.xlabel('Etapas del proceso')
plt.ylabel('Tiempo (minutos)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

for barra, tiempo in zip(barras, tiempos):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 0.3,
             str(tiempo), ha='center', fontsize=11)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos simulados
muestras = list(range(1, 21))
valores = [10, 11, 9, 10, 12, 11, 10, 9, 10, 11, 10, 12, 13, 10, 9, 11, 10, 14, 10, 9]

media = np.mean(valores)
desv = np.std(valores)
lsc = media + 2 * desv
lic = media - 2 * desv

plt.figure(figsize=(12,6))
plt.plot(muestras, valores, marker='o', linewidth=2, label='Proceso')
plt.axhline(media, color='green', linestyle='--', label=f'Media = {media:.2f}')
plt.axhline(lsc, color='red', linestyle='--', label=f'LSC = {lsc:.2f}')
plt.axhline(lic, color='orange', linestyle='--', label=f'LIC = {lic:.2f}')

# Resaltar puntos fuera de control
for i, v in enumerate(valores):
    if v > lsc or v < lic:
        plt.scatter(muestras[i], v, color='red', s=100, zorder=5)

plt.title('Gráfico de Control del Proceso', fontsize=16)
plt.xlabel('Muestra')
plt.ylabel('Valor medido')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -----------------------------
# 1. Datos base del proceso
# -----------------------------
etapas = ['Recepción', 'Picking', 'Packing', 'Facturación', 'Despacho']

# Escenario actual (AS-IS)
asis = [5, 14, 10, 6, 15]

# Escenario 1: mejora continua
mejora_continua = [4, 12, 9, 5, 13]

# Escenario 2: automatización / rediseño
automatizacion = [3, 8, 7, 4, 10]

# -----------------------------
# 2. Gráfica 1: comparación por etapas
# -----------------------------
x = np.arange(len(etapas))
ancho = 0.25

plt.figure(figsize=(12,6))
plt.bar(x - ancho, asis, width=ancho, label='AS-IS')
plt.bar(x, mejora_continua, width=ancho, label='Mejora continua')
plt.bar(x + ancho, automatizacion, width=ancho, label='Automatización')

plt.xticks(x, etapas)
plt.ylabel('Tiempo por etapa (min)')
plt.title('Comparación de escenarios por etapa del proceso')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()

# -----------------------------
# 3. Calcular tiempo total por escenario
# -----------------------------
tiempo_total = [
    sum(asis),
    sum(mejora_continua),
    sum(automatizacion)
]

escenarios = ['AS-IS', 'Mejora continua', 'Automatización']

# Supuestos de costo por pedido
costos = [12.5, 10.8, 9.2]

# Supuestos de errores por cada 100 pedidos
errores = [9, 6, 3]

# -----------------------------
# 4. Crear tabla resumen
# -----------------------------
df = pd.DataFrame({
    'Escenario': escenarios,
    'Tiempo total (min)': tiempo_total,
    'Costo por pedido ($)': costos,
    'Errores por 100 pedidos': errores
})

print("Tabla resumen de escenarios:")
print(df)

# -----------------------------
# 5. Gráfica 2: tiempo total
# -----------------------------
plt.figure(figsize=(9,6))
barras = plt.bar(escenarios, tiempo_total, edgecolor='black')

plt.title('Tiempo total del proceso por escenario')
plt.ylabel('Minutos')
plt.grid(axis='y', linestyle='--', alpha=0.4)

for barra, valor in zip(barras, tiempo_total):
    plt.text(barra.get_x() + barra.get_width()/2, valor + 0.5,
             str(valor), ha='center')

plt.show()

# -----------------------------
# 6. Gráfica 3: costo por pedido
# -----------------------------
plt.figure(figsize=(9,6))
barras = plt.bar(escenarios, costos, edgecolor='black')

plt.title('Costo por pedido por escenario')
plt.ylabel('Costo ($)')
plt.grid(axis='y', linestyle='--', alpha=0.4)

for barra, valor in zip(barras, costos):
    plt.text(barra.get_x() + barra.get_width()/2, valor + 0.1,
             str(valor), ha='center')

plt.show()

# -----------------------------
# 7. Gráfica 4: errores por 100 pedidos
# -----------------------------
plt.figure(figsize=(9,6))
barras = plt.bar(escenarios, errores, edgecolor='black')

plt.title('Errores por cada 100 pedidos')
plt.ylabel('Número de errores')
plt.grid(axis='y', linestyle='--', alpha=0.4)

for barra, valor in zip(barras, errores):
    plt.text(barra.get_x() + barra.get_width()/2, valor + 0.1,
             str(valor), ha='center')

plt.show()

# -----------------------------
# 8. Evaluación simple con puntajes
#    Menor valor = mejor
# -----------------------------
df_eval = df.copy()

# Ranking por criterio: menor tiempo, menor costo y menos errores = mejor
df_eval['Puntaje tiempo'] = df_eval['Tiempo total (min)'].rank(method='min')
df_eval['Puntaje costo'] = df_eval['Costo por pedido ($)'].rank(method='min')
df_eval['Puntaje errores'] = df_eval['Errores por 100 pedidos'].rank(method='min')

df_eval['Puntaje total'] = (
    df_eval['Puntaje tiempo'] +
    df_eval['Puntaje costo'] +
    df_eval['Puntaje errores']
)

print("\nEvaluación de escenarios:")
print(df_eval[['Escenario', 'Puntaje tiempo', 'Puntaje costo', 'Puntaje errores', 'Puntaje total']])

mejor_escenario = df_eval.loc[df_eval['Puntaje total'].idxmin(), 'Escenario']
print(f"\nLa mejor alternativa según el análisis es: {mejor_escenario}")
```
